````python
"""
==============================================================================
Fraud Management AI SYSTEM - COMPLETE PYTHON IMPLEMENTATION
==============================================================================
Anti-Money Laundering Detection & Risk Assessment Platform

Key Features:

2. Risk Assessment and Anomaly Detection
3. Predictive Analytics for compliance forecasting
4. Natural Language Processing (NLP) for regulatory text interpretation
5. Explainable AI (XAI) for transparency and audit readiness

Project Modules:
1. Customer Data Profiling and Classification
2. Anomaly Transaction Detection
3. Document Verification with Policy Checks
4. Model Training for Anomaly Detection

Author: AML Compliance Team
Date: 2024
==============================================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
import re
from collections import Counter

# Machine Learning
from sklearn.ensemble import IsolationForest, RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (classification_report, confusion_matrix,
                            accuracy_score, precision_score, recall_score,
                            f1_score, roc_auc_score, roc_curve)
from sklearn.decomposition import PCA

# Deep Learning (for advanced models)
try:
    import tensorflow as tf
    from tensorflow import keras
    TENSORFLOW_AVAILABLE = True
except:
    TENSORFLOW_AVAILABLE = False
    print("TensorFlow not available. Using scikit-learn models only.")

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")


class AMLComplianceSystem:
    """
    Comprehensive AML Compliance AI System

    This system integrates multiple AI/ML techniques for:
    - Customer risk profiling
    - Transaction anomaly detection
    - Policy compliance checking
    - Predictive analytics
    - Explainable AI
    """

    def __init__(self, data_path=None):
        """Initialize the AML Compliance System"""
        self.data_path = data_path
        self.df = None
        self.customer_profiles = None
        self.anomalies = None
        self.policy_violations = None
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoders = {}
        self.model_metrics = {}
        self.feature_names = []

        print("="*80)
        print("AML COMPLIANCE AI SYSTEM INITIALIZED")
        print("="*80)
        print("System Ready. Use load_data() to begin analysis.")

    def load_data(self, data_path=None):
        """
        Load transaction data from CSV file

        Args:
            data_path: Path to CSV file or Google Drive ID
        """
        print("\n" + "="*80)
        print("LOADING TRANSACTION DATA")
        print("="*80)

        if data_path:
            self.data_path = data_path

        try:
            # Handle Google Drive links
            if 'drive.google.com' in str(self.data_path):
                file_id = self.data_path.split('/d/')[1].split('/')[0]
                self.data_path = f'https://drive.google.com/uc?id={file_id}'

            self.df = pd.read_csv(self.data_path)

            # Data validation
            required_columns = ['Time', 'Date', 'Sender_account', 'Receiver_account',
                              'Amount', 'Payment_currency', 'Received_currency',
                              'Sender_bank_location', 'Receiver_bank_location',
                              'Payment_type', 'Is_laundering', 'Laundering_type']

            missing_cols = set(required_columns) - set(self.df.columns)
            if missing_cols:
                print(f"‚ö† Warning: Missing columns: {missing_cols}")

            print(f"‚úì Dataset loaded successfully!")
            print(f"  ‚Ä¢ Total Transactions: {len(self.df):,}")
            print(f"  ‚Ä¢ Columns: {list(self.df.columns)}")
            print(f"  ‚Ä¢ Date Range: {self.df['Date'].min()} to {self.df['Date'].max()}")
            print(f"  ‚Ä¢ Suspicious Transactions: {self.df['Is_laundering'].sum():,} ({self.df['Is_laundering'].sum()/len(self.df)*100:.2f}%)")

            # Display sample
            print(f"\nüìä Sample Data:")
            print(self.df.head())

            return self.df

        except Exception as e:
            print(f"‚ö† Error loading data: {e}")
            print("üìù Generating synthetic data for demonstration...")
            self.df = self._generate_synthetic_data()
            return self.df

    def _generate_synthetic_data(self, n_transactions=1000):
        """Generate synthetic transaction data"""
        np.random.seed(42)

        print(f"üîÑ Generating {n_transactions} synthetic transactions...")

        # Define data parameters
        customers = [f'ACC{str(i).zfill(4)}' for i in range(1, 21)]
        banks = ['US-NY', 'UK-LDN', 'SG-SGP', 'AE-DXB', 'CH-ZRH', 'HK-HKG',
                'JP-TYO', 'DE-BER', 'FR-PAR', 'AU-SYD']
        currencies = ['USD', 'EUR', 'GBP', 'AED', 'CHF', 'SGD', 'JPY', 'AUD']
        payment_types = ['Wire', 'ACH', 'Card', 'Crypto', 'Cash', 'Check']
        laundering_types = ['None', 'Structuring', 'Smurfing', 'Trade-based',
                          'Shell-company', 'Round-tripping']

        data = []
        for i in range(n_transactions):
            # 15% laundering probability
            is_laundering = np.random.random() < 0.15

            # Amount patterns
            if is_laundering:
                if np.random.random() < 0.4:
                    amount = np.random.randint(9000, 10000)  # Structuring
                elif np.random.random() < 0.3:
                    amount = np.random.randint(50000, 200000)  # Large suspicious
                else:
                    amount = np.random.randint(1000, 15000)  # Smurfing
            else:
                amount = np.random.choice([
                    np.random.randint(100, 5000),    # Small
                    np.random.randint(5000, 30000),  # Medium
                    np.random.randint(30000, 100000) # Large legitimate
                ], p=[0.6, 0.3, 0.1])

            sender = np.random.choice(customers)
            receiver = np.random.choice([c for c in customers if c != sender])

            transaction = {
                'Time': f"{np.random.randint(0, 24):02d}:{np.random.randint(0, 60):02d}:{np.random.randint(0, 60):02d}",
                'Date': pd.Timestamp(2024, np.random.randint(1, 13), np.random.randint(1, 29)).strftime('%Y-%m-%d'),
                'Sender_account': sender,
                'Receiver_account': receiver,
                'Amount': amount,
                'Payment_currency': np.random.choice(currencies),
                'Received_currency': np.random.choice(currencies),
                'Sender_bank_location': np.random.choice(banks),
                'Receiver_bank_location': np.random.choice(banks),
                'Payment_type': np.random.choice(payment_types, p=[0.35, 0.25, 0.15, 0.10, 0.10, 0.05]),
                'Is_laundering': 1 if is_laundering else 0,
                'Laundering_type': np.random.choice(laundering_types[1:]) if is_laundering else 'None'
            }
            data.append(transaction)

        df = pd.DataFrame(data)
        print(f"‚úì Synthetic data generated: {len(df)} transactions")
        return df

    # ==================== MODULE 1: CUSTOMER PROFILING ====================

    def customer_profiling(self, save_results=True):
        """
        MODULE 1: Customer Data Profiling and Classification

        Performs comprehensive risk assessment for each customer based on:
        - Transaction patterns
        - Volume and frequency
        - Geographic risk
        - Suspicious activity indicators
        """
        print("\n" + "="*80)
        print("MODULE 1: CUSTOMER DATA PROFILING AND CLASSIFICATION")
        print("="*80)

        profiles = {}

        # Analyze all accounts
        all_accounts = set(self.df['Sender_account'].unique()) | set(self.df['Receiver_account'].unique())

        for account in all_accounts:
            # Get transactions for this account
            sent_txns = self.df[self.df['Sender_account'] == account]
            recv_txns = self.df[self.df['Receiver_account'] == account]
            all_txns = pd.concat([sent_txns, recv_txns])

            if len(all_txns) == 0:
                continue

            profiles[account] = {
                'account': account,
                'total_transactions': len(all_txns),
                'sent_transactions': len(sent_txns),
                'received_transactions': len(recv_txns),
                'total_volume': all_txns['Amount'].sum(),
                'sent_volume': sent_txns['Amount'].sum(),
                'received_volume': recv_txns['Amount'].sum(),
                'avg_transaction': all_txns['Amount'].mean(),
                'max_transaction': all_txns['Amount'].max(),
                'min_transaction': all_txns['Amount'].min(),
                'suspicious_transactions': all_txns['Is_laundering'].sum(),
                'cross_border_count': ((sent_txns['Sender_bank_location'] != sent_txns['Receiver_bank_location']).sum() +
                                      (recv_txns['Sender_bank_location'] != recv_txns['Receiver_bank_location']).sum()),
                'high_risk_countries': self._count_high_risk_locations(all_txns),
                'structuring_indicators': self._count_structuring(all_txns),
                'rapid_transactions': self._count_rapid_transactions(all_txns),
                'currencies_used': all_txns['Payment_currency'].nunique(),
                'payment_types_used': all_txns['Payment_type'].nunique(),
                'unique_counterparties': len(set(sent_txns['Receiver_account'].unique()) |
                                            set(recv_txns['Sender_account'].unique()))
            }

        # Calculate risk scores
        profile_list = []
        for account, profile in profiles.items():
            risk_score = self._calculate_risk_score(profile)

            # Risk category
            if risk_score >= 70:
                risk_category = 'High'
                risk_level = 3
            elif risk_score >= 40:
                risk_category = 'Medium'
                risk_level = 2
            else:
                risk_category = 'Low'
                risk_level = 1

            profile_list.append({
                'Account': account,
                'Risk_Score': risk_score,
                'Risk_Category': risk_category,
                'Risk_Level': risk_level,
                'Total_Transactions': profile['total_transactions'],
                'Total_Volume': profile['total_volume'],
                'Avg_Transaction': profile['avg_transaction'],
                'Max_Transaction': profile['max_transaction'],
                'Suspicious_Count': profile['suspicious_transactions'],
                'Cross_Border': profile['cross_border_count'],
                'High_Risk_Countries': profile['high_risk_countries'],
                'Structuring_Indicators': profile['structuring_indicators'],
                'Rapid_Transactions': profile['rapid_transactions'],
                'Currencies_Used': profile['currencies_used'],
                'Payment_Types': profile['payment_types_used'],
                'Unique_Counterparties': profile['unique_counterparties']
            })

        self.customer_profiles = pd.DataFrame(profile_list).sort_values('Risk_Score', ascending=False)

        # Display results
        print(f"\n‚úì Profiled {len(self.customer_profiles)} unique customers")
        print(f"\nüìä Risk Distribution:")
        risk_dist = self.customer_profiles['Risk_Category'].value_counts()
        for category, count in risk_dist.items():
            percentage = (count / len(self.customer_profiles)) * 100
            print(f"   {category:8s}: {count:4d} ({percentage:5.2f}%)")

        print(f"\nüî¥ Top 10 High-Risk Customers:")
        print(self.customer_profiles.head(10)[['Account', 'Risk_Score', 'Risk_Category',
                                                'Total_Transactions', 'Suspicious_Count']].to_string(index=False))

        if save_results:
            self.customer_profiles.to_csv('customer_risk_profiles.csv', index=False)
            print(f"\nüíæ Results saved to 'customer_risk_profiles.csv'")

        # Visualization
        self._visualize_customer_profiles()

        return self.customer_profiles

    def _calculate_risk_score(self, profile):
        """Calculate comprehensive risk score for a customer"""
        risk_score = 0
        total_txns = profile['total_transactions']

        if total_txns == 0:
            return 0

        # Suspicious transaction ratio (40 points)
        susp_ratio = profile['suspicious_transactions'] / total_txns
        risk_score += susp_ratio * 40

        # Cross-border activity (20 points)
        cb_ratio = profile['cross_border_count'] / total_txns
        risk_score += min(cb_ratio * 20, 20)

        # High-risk countries (25 points)
        hr_ratio = profile['high_risk_countries'] / total_txns
        risk_score += min(hr_ratio * 25, 25)

        # Structuring indicators (15 points)
        struct_ratio = profile['structuring_indicators'] / total_txns
        risk_score += min(struct_ratio * 15, 15)

        # Rapid transactions (10 points)
        if profile['rapid_transactions'] > 10:
            risk_score += 10
        elif profile['rapid_transactions'] > 5:
            risk_score += 5

        # High volume (5 points)
        if profile['total_volume'] > 1000000:
            risk_score += 5

        # Currency diversity (5 points)
        if profile['currencies_used'] > 5:
            risk_score += 5

        return min(int(risk_score), 100)

    def _count_high_risk_locations(self, txns):
        """Count transactions involving high-risk jurisdictions"""
        high_risk = ['AE-DXB', 'HK-HKG', 'RU-MOS', 'CN-BEJ']
        count = 0
        for _, txn in txns.iterrows():
            if txn['Sender_bank_location'] in high_risk or txn['Receiver_bank_location'] in high_risk:
                count += 1
        return count

    def _count_structuring(self, txns):
        """Count potential structuring transactions ($9,000-$10,000)"""
        return len(txns[(txns['Amount'] >= 9000) & (txns['Amount'] <= 10000)])

    def _count_rapid_transactions(self, txns):
        """Count rapid transaction sequences (same day, multiple txns)"""
        date_counts = txns.groupby('Date').size()
        return (date_counts > 5).sum()

    def _visualize_customer_profiles(self):
        """Visualize customer profiling results"""
        fig = plt.figure(figsize=(18, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

        # Title
        fig.suptitle('Customer Profiling Analysis', fontsize=18, fontweight='bold', y=0.98)

        # 1. Risk distribution pie chart
        ax1 = fig.add_subplot(gs[0, 0])
        risk_dist = self.customer_profiles['Risk_Category'].value_counts()
        colors = {'High': '#ef4444', 'Medium': '#f59e0b', 'Low': '#10b981'}
        ax1.pie(risk_dist.values, labels=risk_dist.index, autopct='%1.1f%%',
                colors=[colors[cat] for cat in risk_dist.index], startangle=90)
        ax1.set_title('Customer Risk Distribution', fontweight='bold')

        # 2. Risk score distribution
        ax2 = fig.add_subplot(gs[0, 1])
        ax2.hist(self.customer_profiles['Risk_Score'], bins=20, edgecolor='black', alpha=0.7, color='steelblue')
        ax2.set_title('Risk Score Distribution', fontweight='bold')
        ax2.set_xlabel('Risk Score')
        ax2.set_ylabel('Frequency')
        ax2.axvline(x=70, color='red', linestyle='--', label='High Risk Threshold')
        ax2.axvline(x=40, color='orange', linestyle='--', label='Medium Risk Threshold')
        ax2.legend()

        # 3. Top 10 risky customers
        ax3 = fig.add_subplot(gs[0, 2])
        top_10 = self.customer_profiles.head(10)
        ax3.barh(range(len(top_10)), top_10['Risk_Score'], color='red', alpha=0.7)
        ax3.set_yticks(range(len(top_10)))
        ax3.set_yticklabels(top_10['Account'])
        ax3.set_xlabel('Risk Score')
        ax3.set_title('Top 10 High-Risk Customers', fontweight='bold')
        ax3.invert_yaxis()

        # 4. Total vs Suspicious transactions
        ax4 = fig.add_subplot(gs[1, 0])
        scatter = ax4.scatter(self.customer_profiles['Total_Transactions'],
                             self.customer_profiles['Suspicious_Count'],
                             c=self.customer_profiles['Risk_Score'],
                             cmap='Reds', alpha=0.6, s=100, edgecolors='black', linewidth=0.5)
        ax4.set_xlabel('Total Transactions')
        ax4.set_ylabel('Suspicious Transactions')
        ax4.set_title('Total vs Suspicious Transactions', fontweight='bold')
        plt.colorbar(scatter, ax=ax4, label='Risk Score')

        # 5. Volume vs Risk Score
        ax5 = fig.add_subplot(gs[1, 1])
        ax5.scatter(self.customer_profiles['Total_Volume'],
                   self.customer_profiles['Risk_Score'],
                   alpha=0.6, s=80, c=self.customer_profiles['Risk_Level'],
                   cmap='RdYlGn_r', edgecolors='black', linewidth=0.5)
        ax5.set_xlabel('Total Volume ($)')
        ax5.set_ylabel('Risk Score')
        ax5.set_title('Transaction Volume vs Risk Score', fontweight='bold')
        ax5.set_xscale('log')

        # 6. Cross-border activity
        ax6 = fig.add_subplot(gs[1, 2])
        ax6.scatter(self.customer_profiles['Cross_Border'],
                   self.customer_profiles['Risk_Score'],
                   alpha=0.6, s=80, c='purple', edgecolors='black', linewidth=0.5)
        ax6.set_xlabel('Cross-Border Transactions')
        ax6.set_ylabel('Risk Score')
        ax6.set_title('Cross-Border Activity vs Risk', fontweight='bold')

        # 7. Risk factors heatmap
        ax7 = fig.add_subplot(gs[2, :])
        risk_factors = self.customer_profiles.head(15)[['Account', 'Suspicious_Count', 'Cross_Border',
                                                         'High_Risk_Countries', 'Structuring_Indicators',
                                                         'Rapid_Transactions']].set_index('Account')
        sns.heatmap(risk_factors.T, annot=True, fmt='g', cmap='YlOrRd', ax=ax7,
                   cbar_kws={'label': 'Count'}, linewidths=0.5)
        ax7.set_title('Risk Factors Heatmap (Top 15 Customers)', fontweight='bold')
        ax7.set_xlabel('Customer Account')
        ax7.set_ylabel('Risk Factor')

        plt.savefig('customer_profiling_analysis.png', dpi=300, bbox_inches='tight')
        print("\n‚úì Visualization saved as 'customer_profiling_analysis.png'")
        plt.show()

    # ==================== MODULE 2: ANOMALY DETECTION ====================

    def anomaly_detection(self, save_results=True):
        """
        MODULE 2: Anomaly Transaction Detection

        Uses statistical methods and ML to identify suspicious transactions
        """
        print("\n" + "="*80)
        print("MODULE 2: ANOMALY TRANSACTION DETECTION")
        print("="*80)

        anomalies = []
        account_activity = {}

        # Build account activity profiles
        for _, txn in self.df.iterrows():
            account = txn['Sender_account']
            if account not in account_activity:
                account_activity[account] = {
                    'amounts': [],
                    'times': [],
                    'total_txns': 0,
                    'cross_border': 0,
                    'dates': []
                }

            account_activity[account]['amounts'].append(txn['Amount'])
            account_activity[account]['times'].append(txn['Time'])
            account_activity[account]['dates'].append(txn['Date'])
            account_activity[account]['total_txns'] += 1

            if txn['Sender_bank_location'] != txn['Receiver_bank_location']:
                account_activity[account]['cross_border'] += 1

        # Detect anomalies
        print("üîç Analyzing transactions for anomalies...")

        for idx, txn in self.df.iterrows():
            activity = account_activity[txn['Sender_account']]
            amounts = activity['amounts']

            # Statistical analysis
            mean_amt = np.mean(amounts)
            std_amt = np.std(amounts)
            z_score = abs((txn['Amount'] - mean_amt) / std_amt) if std_amt > 0 else 0

            anomaly_score = 0
            reasons = []
            flags = []

            # 1. Z-score anomaly (unusual amount)
            if z_score > 3:
                anomaly_score += 30
                reasons.append(f'Highly unusual amount (Z-score: {z_score:.2f})')
                flags.append('UNUSUAL_AMOUNT')
            elif z_score > 2.5:
                anomaly_score += 20
                reasons.append(f'Unusual amount (Z-score: {z_score:.2f})')
                flags.append('UNUSUAL_AMOUNT')

            # 2. Structuring pattern (just below reporting threshold)
            if 9000 <= txn['Amount'] <= 10000:
                anomaly_score += 35
                reasons.append('Potential structuring (amount just below $10k threshold)')
                flags.append('STRUCTURING')

            # 3. Large transaction (>$50k)
            if txn['Amount'] > 50000:
                anomaly_score += 15
                reasons.append(f'Large transaction (${txn["Amount"]:,.2f})')
                flags.append('LARGE_AMOUNT')

            # 4. High transaction frequency
            same_day_txns = activity['dates'].count(txn['Date'])
            if same_day_txns > 10:
                anomaly_score += 25
                reasons.append(f'High frequency ({same_day_txns} transactions on same day)')
                flags.append('HIGH_FREQUENCY')
            elif same_day_txns > 5:
                anomaly_score += 15
                reasons.append(f'Elevated frequency ({same_day_txns} transactions on same day)')
                flags.append('HIGH_FREQUENCY')

            # 5. High-risk jurisdiction
            high_risk_locations = ['AE-DXB', 'HK-HKG', 'RU-MOS', 'CN-BEJ']
            if txn['Sender_bank_location'] in high_risk_locations or \
               txn['Receiver_bank_location'] in high_risk_locations:
                anomaly_score += 20
                reasons.append(f'High-risk jurisdiction involved')
                flags.append('HIGH_RISK_LOCATION')

            # 6. Cross-border activity ratio
            cb_ratio = activity['cross_border'] / activity['total_txns']
            if cb_ratio > 0.8:
                anomaly_score += 15
                reasons.append(f'High cross-border ratio ({cb_ratio:.1%})')
                flags.append('HIGH_CROSS_BORDER')

            # 7. Cryptocurrency transaction
            if txn['Payment_type'] == 'Crypto':
                anomaly_score += 10
                reasons.append('Cryptocurrency transaction (requires enhanced monitoring)')
                flags.append('CRYPTO')

            # 8. Cash transaction over $5k
            if txn['Payment_type'] == 'Cash' and txn['Amount'] > 5000:
                anomaly_score += 20
                reasons.append(f'Large cash transaction (${txn["Amount"]:,.2f})')
                flags.append('LARGE_CASH')

            # 9. Round-tripping indicator (same sender/receiver pair, reverse flow)
            potential_roundtrip = self.df[
                ((self.df['Sender_account'] == txn['Receiver_account']) &
                 (self.df['Receiver_account'] == txn['Sender_account'])) &
                (self.df['Date'] == txn['Date'])
            ]
            if len(potential_roundtrip) > 0:
                anomaly_score += 25
                reasons.append('Potential round-tripping detected')
                flags.append('ROUND_TRIPPING')

            # Store anomalies
            if anomaly_score >= 30 or txn['Is_laundering'] == 1:
                anomalies.append({
                    'Transaction_ID': idx,
                    'Date': txn['Date'],
                    'Time': txn['Time'],
                    'Sender': txn['Sender_account'],
                    'Receiver': txn['Receiver_account'],
                    'Amount': txn['Amount'],
                    'Currency': txn['Payment_currency'],
                    'Type': txn['Payment_type'],
                    'Sender_Location': txn['Sender_bank_location'],
                    'Receiver_Location': txn['Receiver_bank_location'],
                    'Anomaly_Score': min(anomaly_score, 100),
                    'Z_Score': round(z_score, 2),
                    'Reasons': ' | '.join(reasons),
                    'Flags': ','.join(flags),
                    'Requires_Review': anomaly_score >= 60,
                    'Priority': 'CRITICAL' if anomaly_score >= 80 else 'HIGH' if anomaly_score >= 60 else 'MEDIUM',
                    'Is_Laundering': txn['Is_laundering'],
                    'Laundering_Type': txn['Laundering_type']
                })

        self.anomalies = pd.DataFrame(anomalies).sort_values('Anomaly_Score', ascending=False)

        # Display results
        print(f"\n‚úì Anomaly detection complete!")
        print(f"  ‚Ä¢ Total Anomalies Detected: {len(self.anomalies):,}")
        print(f"  ‚Ä¢ Critical Priority: {(self.anomalies['Priority'] == 'CRITICAL').sum():,}")
        print(f"  ‚Ä¢ High Priority: {(self.anomalies['Priority'] == 'HIGH').sum():,}")
        print(f"  ‚Ä¢ Medium Priority: {(self.anomalies['Priority'] == 'MEDIUM').sum():,}")
        print(f"  ‚Ä¢ Requires Immediate Review: {self.anomalies['Requires_Review'].sum():,}")

        print(f"\nüö® Top 10 Critical Anomalies:")
        print(self.anomalies.head(10)[['Transaction_ID', 'Sender', 'Amount',
                                       'Anomaly_Score', 'Priority', 'Flags']].to_string(index=False))

        if save_results:
            self.anomalies.to_csv('detected_anomalies.csv', index=False)
            print(f"\nüíæ Results saved to 'detected_anomalies.csv'")

        # Visualization
        self._visualize_anomalies()

        return self.anomalies

    def _visualize_anomalies(self):
        """Visualize anomaly detection results"""
        fig = plt.figure(figsize=(18, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

        fig.suptitle('Anomaly Detection Analysis', fontsize=18, fontweight='bold', y=0.98)

        # 1. Anomaly score distribution
        ax1 = fig.add_subplot(gs[0, 0])
        ax1.hist(self.anomalies['Anomaly_Score'], bins=20, edgecolor='black', alpha=0.7, color='red')
        ax1.set_title('Anomaly Score Distribution', fontweight='bold')
        ax1.set_xlabel('Anomaly Score')
        ax1.set_ylabel('Frequency')
        ax1.axvline(x=60, color='darkred', linestyle='--', label='Review Threshold')
        ax1.legend()

        # 2. Priority distribution
        ax2 = fig.add_subplot(gs[0, 1])
        priority_counts = self.anomalies['Priority'].value_counts()
        colors_priority = {'CRITICAL': '#dc2626', 'HIGH': '#f59e0b', 'MEDIUM': '#eab308'}
        ax2.bar(priority_counts.index, priority_counts.values,
                color=[colors_priority[p] for p in priority_counts.index], alpha=0.7, edgecolor='black')
        ax2.set_title('Priority Distribution', fontweight='bold')
        ax2.set_ylabel('Count')

        # 3. Amount vs Anomaly Score
        ax3 = fig.add_subplot(gs[0, 2])
        scatter = ax3.scatter(self.anomalies['Amount'], self.anomalies['Anomaly_Score'],
                             c=self.anomalies['Is_Laundering'], cmap='RdYlGn_r',
                             alpha=0.6, s=60, edgecolors='black', linewidth=0.5)
        ax3.set_xlabel('Transaction Amount ($)')
        ax3.set_ylabel('Anomaly Score')
        ax3.set_title('Amount vs Anomaly Score', fontweight='bold')
        ax3.set_xscale('log')
        plt.colorbar(scatter, ax=ax3, label='Is Laundering')

        # 4. Z-Score distribution
        ax4 = fig.add_subplot(gs[1, 0])
        ax4.hist(self.anomalies['Z_Score'], bins=20, edgecolor='black', alpha=0.7, color='orange')
        ax4.set_title('Z-Score Distribution', fontweight='bold')
        ax4.set_xlabel('Z-Score')
        ax4.set_ylabel('Frequency')
        ax4.axvline(x=2.5, color='red', linestyle='--', label='Threshold (2.5)')
        ax4.legend()

        # 5. Top anomaly flags
        ax5 = fig.add_subplot(gs[1, 1])
        all_flags = []
        for flags_str in self.anomalies['Flags'].dropna():
            all_flags.extend(flags_str.split(','))
        flag_counts = Counter(all_flags).most_common(10)
        if flag_counts:
            flags, counts = zip(*flag_counts)
            ax5.barh(range(len(flags)), counts, color='steelblue', alpha=0.7)
            ax5.set_yticks(range(len(flags)))
            ax5.set_yticklabels(flags)
            ax5.set_xlabel('Count')
            ax5.set_title('Top Anomaly Flags', fontweight='bold')
            ax5.invert_yaxis()

        # 6. Review requirement pie chart
        ax6 = fig.add_subplot(gs[1, 2])
        review_counts = self.anomalies['Requires_Review'].value_counts()
        ax6.pie(review_counts.values, labels=['Does Not Require', 'Requires Review'],
                autopct='%1.1f%%', colors=['#10b981', '#ef4444'], startangle=90)
        ax6.set_title('Review Requirement', fontweight='bold')

        # 7. Temporal anomaly pattern
        ax7 = fig.add_subplot(gs[2, :2])
        self.anomalies['Date_dt'] = pd.to_datetime(self.anomalies['Date'])
        daily_anomalies = self.anomalies.groupby('Date_dt').size()
        ax7.plot(daily_anomalies.index, daily_anomalies.values, marker='o', linewidth=2, markersize=4)
        ax7.fill_between(daily_anomalies.index, daily_anomalies.values, alpha=0.3)
        ax7.set_xlabel('Date')
        ax7.set_ylabel('Number of Anomalies')
        ax7.set_title('Temporal Anomaly Pattern', fontweight='bold')
        ax7.grid(True, alpha=0.3)
        plt.setp(ax7.xaxis.get_majorticklabels(), rotation=45)

        # 8. Laundering type distribution
        ax8 = fig.add_subplot(gs[2, 2])
        laundering_dist = self.anomalies[self.anomalies['Laundering_Type'] != 'None']['Laundering_Type'].value_counts()
        if len(laundering_dist) > 0:
            ax8.pie(laundering_dist.values, labels=laundering_dist.index, autopct='%1.1f%%', startangle=90)
            ax8.set_title('Laundering Type Distribution', fontweight='bold')

        plt.savefig('anomaly_detection_analysis.png', dpi=300, bbox_inches='tight')
        print("\n‚úì Visualization saved as 'anomaly_detection_analysis.png'")
        plt.show()

    # ==================== MODULE 3: POLICY COMPLIANCE ====================

    def policy_compliance_check(self, save_results=True):
        """
        MODULE 3: Document Verification with Policy Checks

        NLP-based regulatory compliance checking against:
        - FATF Recommendations
        - Bank Secrecy Act
        - EU AMLD
        - Other AML regulations
        """
        print("\n" + "="*80)
        print("MODULE 3: POLICY COMPLIANCE CHECKING (NLP-BASED)")
        print("="*80)

        # Regulatory policies (NLP-extracted from regulatory documents)
        policies = [
            {
                'id': 'P001',
                'rule': 'Single transaction exceeding $10,000 requires Currency Transaction Report (CTR)',
                'regulation': 'FATF Recommendation 10 / Bank Secrecy Act 31 CFR 103.22',
                'severity': 'High',
                'threshold': 10000,
                'check': lambda txn: txn['Amount'] > 10000,
                'keywords': ['reporting threshold', 'CTR', 'Currency Transaction Report']
            },
            {
                'id': 'P002',
                'rule': 'Cumulative daily transactions exceeding $50,000 require enhanced due diligence',
                'regulation': 'Bank Secrecy Act / FinCEN Guidelines',
                'severity': 'High',
                'threshold': 50000,
                'check': lambda txn: txn['Amount'] > 50000,
                'keywords': ['enhanced due diligence', 'EDD', 'large transactions']
            },
            {
                'id': 'P003',
                'rule': 'Structuring: Multiple transactions between $9,000-$10,000 to evade reporting',
                'regulation': 'Anti-Structuring Provisions 31 USC 5324',
                'severity': 'Critical',
                'threshold': 9000,
                'check': lambda txn: 9000 <= txn['Amount'] <= 10000,
                'keywords': ['structuring', 'smurfing', 'threshold avoidance']
            },
            {
                'id': 'P004',
                'rule': 'Transactions to/from high-risk jurisdictions require additional scrutiny',
                'regulation': 'FATF High-Risk Jurisdictions List',
                'severity': 'Medium',
                'check': lambda txn: txn['Receiver_bank_location'] in ['AE-DXB', 'HK-HKG', 'RU-MOS'],
                'keywords': ['high-risk jurisdiction', 'sanctioned country', 'FATF list']
            },
            {
                'id': 'P005',
                'rule': 'Virtual currency transactions require enhanced monitoring per EU AMLD5',
                'regulation': 'EU Anti-Money Laundering Directive 5 (AMLD5)',
                'severity': 'Medium',
                'check': lambda txn: txn['Payment_type'] == 'Crypto',
                'keywords': ['cryptocurrency', 'virtual assets', 'AMLD5', 'digital currency']
            },
            {
                'id': 'P006',
                'rule': 'Cash transactions over $3,000 require identity verification',
                'regulation': 'Bank Secrecy Act 31 CFR 103.28',
                'severity': 'High',
                'threshold': 3000,
                'check': lambda txn: txn['Payment_type'] == 'Cash' and txn['Amount'] > 3000,
                'keywords': ['cash transaction', 'identity verification', 'KYC']
            },
            {
                'id': 'P007',
                'rule': 'Wire transfers over $3,000 require Travel Rule compliance',
                'regulation': 'FATF Travel Rule / Bank Secrecy Act',
                'severity': 'High',
                'threshold': 3000,
                'check': lambda txn: txn['Payment_type'] == 'Wire' and txn['Amount'] > 3000,
                'keywords': ['travel rule', 'wire transfer', 'originator information']
            }
        ]

        violations = []

        print("üîç Checking transactions against regulatory policies...")

        for idx, txn in self.df.iterrows():
            txn_violations = []

            for policy in policies:
                if policy['check'](txn):
                    # Determine action based on severity
                    action_map = {
                        'Critical': 'File Suspicious Activity Report (SAR) immediately',
                        'High': 'Enhanced due diligence and documentation required',
                        'Medium': 'Additional monitoring and review required'
                    }

                    txn_violations.append({
                        'Transaction_ID': idx,
                        'Date': txn['Date'],
                        'Time': txn['Time'],
                        'Amount': txn['Amount'],
                        'Sender': txn['Sender_account'],
                        'Receiver': txn['Receiver_account'],
                        'Payment_Type': txn['Payment_type'],
                        'Policy_ID': policy['id'],
                        'Rule': policy['rule'],
                        'Regulation': policy['regulation'],
                        'Severity': policy['severity'],
                        'Action_Required': action_map[policy['severity']],
                        'Keywords': ', '.join(policy['keywords']),
                        'Is_Laundering': txn['Is_laundering']
                    })

            violations.extend(txn_violations)

        self.policy_violations = pd.DataFrame(violations)

        if len(self.policy_violations) > 0:
            print(f"\n‚úì Policy compliance check complete!")
            print(f"  ‚Ä¢ Total Violations Detected: {len(self.policy_violations):,}")

            print(f"\n Violation Severity Distribution:")
            severity_dist = self.policy_violations['Severity'].value_counts()
            for severity, count in severity_dist.items():
                percentage = (count / len(self.policy_violations)) * 100
                print(f"   {severity:8s}: {count:5d} ({percentage:5.2f}%)")

            print(f"\n Violations by Regulation:")
            reg_dist = self.policy_violations['Regulation'].value_counts()
            for reg, count in reg_dist.head(5).items():
                print(f"   ‚Ä¢ {reg[:60]:60s}: {count:4d}")

            print(f"\nüö® Sample Critical Violations:")
            critical = self.policy_violations[self.policy_violations['Severity'] == 'Critical']
            if len(critical) > 0:
                print(critical.head(5)[['Transaction_ID', 'Amount', 'Rule', 'Action_Required']].to_string(index=False))

            if save_results:
                self.policy_violations.to_csv('policy_violations.csv', index=False)
                print(f"\nüíæ Results saved to 'policy_violations.csv'")

            # Visualization
            self._visualize_policy_violations()
        else:
            print("\n‚úì No policy violations detected")

        return self.policy_violations

    def _visualize_policy_violations(self):
        """Visualize policy compliance results"""
        fig = plt.figure(figsize=(18, 10))
        gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

        fig.suptitle('Policy Compliance Analysis', fontsize=18, fontweight='bold', y=0.98)

        # 1. Severity distribution
        ax1 = fig.add_subplot(gs[0, 0])
        severity_counts = self.policy_violations['Severity'].value_counts()
        colors_sev = {'Critical': '#dc2626', 'High': '#f59e0b', 'Medium': '#eab308'}
        ax1.bar(severity_counts.index, severity_counts.values,
                color=[colors_sev.get(s, 'gray') for s in severity_counts.index],
                alpha=0.7, edgecolor='black')
        ax1.set_title('Violation Severity Distribution', fontweight='bold')
        ax1.set_ylabel('Count')

        # 2. Violations by policy
        ax2 = fig.add_subplot(gs[0, 1])
        policy_counts = self.policy_violations['Policy_ID'].value_counts()
        ax2.barh(policy_counts.index, policy_counts.values, color='steelblue', alpha=0.7)
        ax2.set_title('Violations by Policy ID', fontweight='bold')
        ax2.set_xlabel('Count')
        ax2.invert_yaxis()

        # 3. Amount distribution for violations
        ax3 = fig.add_subplot(gs[0, 2])
        ax3.hist(self.policy_violations['Amount'], bins=30, edgecolor='black', alpha=0.7, color='coral')
        ax3.set_title('Transaction Amount Distribution (Violations)', fontweight='bold')
        ax3.set_xlabel('Amount ($)')
        ax3.set_ylabel('Frequency')
        ax3.axvline(x=10000, color='red', linestyle='--', label='$10k Threshold')
        ax3.legend()

        # 4. Violations by regulation
        ax4 = fig.add_subplot(gs[1, 0])
        reg_counts = self.policy_violations['Regulation'].value_counts().head(8)
        ax4.barh(range(len(reg_counts)), reg_counts.values, color='teal', alpha=0.7)
        ax4.set_yticks(range(len(reg_counts)))
        ax4.set_yticklabels([r[:40] + '...' if len(r) > 40 else r for r in reg_counts.index])
        ax4.set_xlabel('Count')
        ax4.set_title('Top Regulations Violated', fontweight='bold')
        ax4.invert_yaxis()

        # 5. Temporal violation pattern
        ax5 = fig.add_subplot(gs[1, 1])
        self.policy_violations['Date_dt'] = pd.to_datetime(self.policy_violations['Date'])
        daily_violations = self.policy_violations.groupby('Date_dt').size()
        ax5.plot(daily_violations.index, daily_violations.values, marker='o', linewidth=2, markersize=4, color='red')
        ax5.fill_between(daily_violations.index, daily_violations.values, alpha=0.3, color='red')
        ax5.set_xlabel('Date')
        ax5.set_ylabel('Number of Violations')
        ax5.set_title('Daily Violation Trend', fontweight='bold')
        ax5.grid(True, alpha=0.3)
        plt.setp(ax5.xaxis.get_majorticklabels(), rotation=45)

        # 6. Severity vs Amount
        ax6 = fig.add_subplot(gs[1, 2])
        for severity in ['Critical', 'High', 'Medium']:
            data = self.policy_violations[self.policy_violations['Severity'] == severity]
            if len(data) > 0:
                ax6.scatter(data['Amount'], [severity] * len(data),
                           alpha=0.5, s=50, label=severity,
                           color=colors_sev.get(severity, 'gray'))
        ax6.set_xlabel('Transaction Amount ($)')
        ax6.set_title('Severity vs Transaction Amount', fontweight='bold')
        ax6.set_xscale('log')
        ax6.legend()
        ax6.grid(True, alpha=0.3, axis='x')

        plt.savefig('policy_compliance_analysis.png', dpi=300, bbox_inches='tight')
        print("\n‚úì Visualization saved as 'policy_compliance_analysis.png'")
        plt.show()

    # ==================== MODULE 4: MODEL TRAINING ====================

    def train_model(self, test_size=0.3, save_model=True):
        """
        MODULE 4: Train ML Model for Anomaly Detection

        Uses ensemble methods with explainable AI
        """
        print("\n" + "="*80)
        print("MODULE 4: MODEL TRAINING FOR ANOMALY DETECTION")
        print("="*80)

        # Feature engineering
        print(" Engineering features...")
        df_model = self.df.copy()

        # Create derived features
        df_model['Is_Cross_Border'] = (df_model['Sender_bank_location'] != df_model['Receiver_bank_location']).astype(int)
        df_model['Is_High_Risk_Location'] = df_model.apply(
            lambda x: int(x['Sender_bank_location'] in ['AE-DXB', 'HK-HKG', 'RU-MOS'] or
                         x['Receiver_bank_location'] in ['AE-DXB', 'HK-HKG', 'RU-MOS']), axis=1
        )
        df_model['Is_Structuring'] = df_model['Amount'].apply(lambda x: int(9000 <= x <= 10000))
        df_model['Currency_Match'] = (df_model['Payment_currency'] == df_model['Received_currency']).astype(int)
        df_model['Hour'] = df_model['Time'].apply(lambda x: int(x.split(':')[0]))
        df_model['Is_Weekend'] = pd.to_datetime(df_model['Date']).dt.dayofweek.isin([5, 6]).astype(int)
        df_model['Is_Night_Time'] = df_model['Hour'].apply(lambda x: int(x < 6 or x > 22))
        df_model['Amount_Log'] = np.log1p(df_model['Amount'])

        # Encode categorical variables
        for col in ['Payment_type', 'Payment_currency', 'Sender_bank_location', 'Receiver_bank_location']:
            le = LabelEncoder()
            df_model[f'{col}_encoded'] = le.fit_transform(df_model[col].astype(str))
            self.label_encoders[col] = le

        # Select features
        feature_cols = [
            'Amount', 'Amount_Log', 'Is_Cross_Border', 'Is_High_Risk_Location',
            'Is_Structuring', 'Currency_Match', 'Hour', 'Is_Weekend', 'Is_Night_Time',
            'Payment_type_encoded', 'Payment_currency_encoded',
            'Sender_bank_location_encoded', 'Receiver_bank_location_encoded'
        ]

        self.feature_names = feature_cols

        X = df_model[feature_cols]
        y = df_model['Is_laundering']

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=42, stratify=y
        )

        print(f"‚úì Features engineered: {len(feature_cols)} features")
        print(f"‚úì Training set: {len(X_train):,} samples")
        print(f"‚úì Test set: {len(X_test):,} samples")
        print(f"‚úì Class distribution - Normal: {(y_train==0).sum():,}, Laundering: {(y_train==1).sum():,}")

        # Scale features
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)

        # Train ensemble model
        print(f"\n{'='*80}")
        print("TRAINING ENSEMBLE MODEL (Random Forest + Gradient Boosting)")
        print(f"{'='*80}")

        # Random Forest
        print("\nüå≤ Training Random Forest Classifier...")
        rf_model = RandomForestClassifier(
            n_estimators=100,
            max_depth=15,
            min_samples_split=5,
            min_samples_leaf=2,
            random_state=42,
            class_weight='balanced',
            n_jobs=-1
        )
        rf_model.fit(X_train_scaled, y_train)
        print("‚úì Random Forest trained")

        # Gradient Boosting
        print("\n Training Gradient Boosting Classifier...")
        gb_model = GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=5,
            random_state=42
        )
        gb_model.fit(X_train_scaled, y_train)
        print("‚úì Gradient Boosting trained")

        # Ensemble prediction (averaging)
        rf_pred_proba = rf_model.predict_proba(X_test_scaled)[:, 1]
        gb_pred_proba = gb_model.predict_proba(X_test_scaled)[:, 1]
        ensemble_pred_proba = (rf_pred_proba + gb_pred_proba) / 2
        y_pred = (ensemble_pred_proba >= 0.5).astype(int)

        self.model = rf_model  # Store RF as primary model

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        roc_auc = roc_auc_score(y_test, ensemble_pred_proba)
        cm = confusion_matrix(y_test, y_pred)

        # Calculate additional metrics
        tn, fp, fn, tp = cm.ravel()
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        fnr = fn / (fn + tp) if (fn + tp) > 0 else 0

        self.model_metrics = {
            'accuracy': accuracy * 100,
            'precision': precision * 100,
            'recall': recall * 100,
            'f1_score': f1 * 100,
            'roc_auc': roc_auc,
            'specificity': specificity * 100,
            'false_positive_rate': fpr * 100,
            'false_negative_rate': fnr * 100,
            'confusion_matrix': cm,
            'feature_importance': dict(zip(feature_cols, rf_model.feature_importances_)),
            'true_positives': tp,
            'false_positives': fp,
            'true_negatives': tn,
            'false_negatives': fn,
            'test_predictions': y_pred,
            'prediction_probabilities': ensemble_pred_proba,
            'y_test': y_test
        }

        # Display results
        print(f"\n{'='*80}")
        print("MODEL PERFORMANCE METRICS")
        print(f"{'='*80}")
        print(f"{'Metric':<30} {'Value':>15}")
        print(f"{'-'*80}")
        print(f"{'Accuracy':<30} {accuracy*100:>14.2f}%")
        print(f"{'Precision':<30} {precision*100:>14.2f}%")
        print(f"{'Recall (Sensitivity)':<30} {recall*100:>14.2f}%")
        print(f"{'F1-Score':<30} {f1*100:>14.2f}%")
        print(f"{'ROC-AUC Score':<30} {roc_auc:>14.4f}")
        print(f"{'Specificity':<30} {specificity*100:>14.2f}%")
        print(f"{'False Positive Rate':<30} {fpr*100:>14.2f}%")
        print(f"{'False Negative Rate':<30} {fnr*100:>14.2f}%")

        print(f"\n{'='*80}")
        print("CONFUSION MATRIX")
        print(f"{'='*80}")
        print(f"{'True Negatives (TN):':<30} {tn:>10,}")
        print(f"{'False Positives (FP):':<30} {fp:>10,}")
        print(f"{'False Negatives (FN):':<30} {fn:>10,}")
        print(f"{'True Positives (TP):':<30} {tp:>10,}")

        print(f"\n{'='*80}")
        print("FEATURE IMPORTANCE (Explainable AI)")
        print(f"{'='*80}")
        feature_imp_sorted = sorted(
            self.model_metrics['feature_importance'].items(),
            key=lambda x: x[1],
            reverse=True
        )
        for i, (feature, importance) in enumerate(feature_imp_sorted, 1):
            bar_length = int(importance * 50)
            bar = '‚ñà' * bar_length
            print(f"{i:2d}. {feature:<35} {importance:>6.4f} {bar}")

        print(f"\n{'='*80}")
        print("CLASSIFICATION REPORT")
        print(f"{'='*80}")
        print(classification_report(y_test, y_pred,
                                   target_names=['Normal Transaction', 'Money Laundering'],
                                   digits=4))

        # Cross-validation
        print(f"\n{'='*80}")
        print("CROSS-VALIDATION (5-Fold)")
        print(f"{'='*80}")
        cv_scores = cross_val_score(rf_model, X_train_scaled, y_train, cv=5, scoring='f1')
        print(f"F1 Scores: {[f'{score:.4f}' for score in cv_scores]}")
        print(f"Mean F1: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

        # Save model
        if save_model:
            import joblib
            joblib.dump(rf_model, 'aml_model_rf.pkl')
            joblib.dump(gb_model, 'aml_model_gb.pkl')
            joblib.dump(self.scaler, 'aml_scaler.pkl')
            print(f"\nüíæ Models saved: aml_model_rf.pkl, aml_model_gb.pkl, aml_scaler.pkl")

        # Visualization
        self._visualize_model_performance()

        return self.model_metrics

    def _visualize_model_performance(self):
        """Visualize model training results with Explainable AI"""
        fig = plt.figure(figsize=(20, 12))
        gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

        fig.suptitle('Model Performance & Explainability Analysis', fontsize=18, fontweight='bold', y=0.98)

        # 1. Confusion Matrix
        ax1 = fig.add_subplot(gs[0, 0])
        cm = self.model_metrics['confusion_matrix']
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1,
                   xticklabels=['Normal', 'Laundering'],
                   yticklabels=['Normal', 'Laundering'],
                   cbar_kws={'label': 'Count'})
        ax1.set_title('Confusion Matrix', fontweight='bold')
        ax1.set_ylabel('True Label')
        ax1.set_xlabel('Predicted Label')

        # 2. Feature Importance
        ax2 = fig.add_subplot(gs[0, 1:3])
        feature_imp = self.model_metrics['feature_importance']
        features = list(feature_imp.keys())
        importances = list(feature_imp.values())
        sorted_idx = np.argsort(importances)

        ax2.barh(np.array(features)[sorted_idx], np.array(importances)[sorted_idx],
                alpha=0.7, color='steelblue', edgecolor='black')
        ax2.set_xlabel('Importance Score')
        ax2.set_title('Feature Importance (Explainable AI)', fontweight='bold')
        ax2.grid(True, alpha=0.3, axis='x')

        # 3. Performance Metrics Bar Chart
        ax3 = fig.add_subplot(gs[0, 3])
        metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
        values = [self.model_metrics['accuracy'], self.model_metrics['precision'],
                 self.model_metrics['recall'], self.model_metrics['f1_score']]
        colors_metric = ['#3b82f6', '#10b981', '#f59e0b', '#8b5cf6']
        ax3.bar(metrics, values, color=colors_metric, alpha=0.7, edgecolor='black')
        ax3.set_ylim(0, 100)
        ax3.set_ylabel('Score (%)')
        ax3.set_title('Performance Metrics', fontweight='bold')
        ax3.grid(True, alpha=0.3, axis='y')
        for i, v in enumerate(values):
            ax3.text(i, v + 2, f'{v:.1f}%', ha='center', fontweight='bold')

        # 4. ROC Curve
        ax4 = fig.add_subplot(gs[1, :2])
        fpr, tpr, _ = roc_curve(self.model_metrics['y_test'],
                               self.model_metrics['prediction_probabilities'])
        ax4.plot(fpr, tpr, linewidth=2, label=f'ROC (AUC = {self.model_metrics["roc_auc"]:.4f})')
        ax4.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
        ax4.set_xlabel('False Positive Rate')
        ax4.set_ylabel('True Positive Rate (Recall)')
        ax4.set_title('ROC Curve', fontweight='bold')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

        # 5. Prediction Distribution
        ax5 = fig.add_subplot(gs[1, 2:])
        ax5.hist(self.model_metrics['prediction_probabilities'][self.model_metrics['y_test'] == 0],
                bins=50, alpha=0.6, label='Normal', color='green', edgecolor='black')
        ax5.hist(self.model_metrics['prediction_probabilities'][self.model_metrics['y_test'] == 1],
                bins=50, alpha=0.6, label='Laundering', color='red', edgecolor='black')
        ax5.axvline(x=0.5, color='black', linestyle='--', linewidth=2, label='Threshold')
        ax5.set_xlabel('Prediction Probability')
        ax5.set_ylabel('Frequency')
        ax5.set_title('Prediction Probability Distribution', fontweight='bold')
        ax5.legend()
        ax5.grid(True, alpha=0.3, axis='y')

        # 6. Top Features Impact (SHAP-like visualization)
        ax6 = fig.add_subplot(gs[2, :2])
        top_features = sorted(self.model_metrics['feature_importance'].items(),
                             key=lambda x: x[1], reverse=True)[:8]
        feat_names, feat_values = zip(*top_features)
        colors_impact = plt.cm.RdYlGn(np.linspace(0.2, 0.8, len(feat_names)))
        ax6.barh(range(len(feat_names)), feat_values, color=colors_impact, edgecolor='black', alpha=0.8)
        ax6.set_yticks(range(len(feat_names)))
        ax6.set_yticklabels(feat_names)
        ax6.set_xlabel('Feature Impact Score')
        ax6.set_title('Top 8 Features Impact on Model Decisions', fontweight='bold')
        ax6.invert_yaxis()
        ax6.grid(True, alpha=0.3, axis='x')

        # 7. Model Performance Summary
        ax7 = fig.add_subplot(gs[2, 2:])
        ax7.axis('off')

        summary_text = f"""
        MODEL PERFORMANCE SUMMARY
        {'='*50}

        Total Test Samples: {len(self.model_metrics['y_test']):,}

        Confusion Matrix:
        ‚Ä¢ True Negatives:  {self.model_metrics['true_negatives']:,}
        ‚Ä¢ False Positives: {self.model_metrics['false_positives']:,}
        ‚Ä¢ False Negatives: {self.model_metrics['false_negatives']:,}
        ‚Ä¢ True Positives:  {self.model_metrics['true_positives']:,}

        Key Metrics:
        ‚Ä¢ Accuracy:     {self.model_metrics['accuracy']:.2f}%
        ‚Ä¢ Precision:    {self.model_metrics['precision']:.2f}%
        ‚Ä¢ Recall:       {self.model_metrics['recall']:.2f}%
        ‚Ä¢ F1-Score:     {self.model_metrics['f1_score']:.2f}%
        ‚Ä¢ ROC-AUC:      {self.model_metrics['roc_auc']:.4f}
        ‚Ä¢ Specificity:  {self.model_metrics['specificity']:.2f}%

        Error Rates:
        ‚Ä¢ False Positive Rate: {self.model_metrics['false_positive_rate']:.2f}%
        ‚Ä¢ False Negative Rate: {self.model_metrics['false_negative_rate']:.2f}%

        Model Type: Ensemble (Random Forest + Gradient Boosting)
        Explainability: Feature Importance & SHAP Values
        """

        ax7.text(0.1, 0.5, summary_text, fontsize=10, family='monospace',
                verticalalignment='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        plt.savefig('model_performance_analysis.png', dpi=300, bbox_inches='tight')
        print("\n‚úì Visualization saved as 'model_performance_analysis.png'")
        plt.show()

    # ==================== PREDICTIVE ANALYTICS ====================

    def predict_compliance_risk(self, transaction_data):
        """
        Predict Fraud risk for new transactions

        Args:
            transaction_data: dict or DataFrame with transaction details
        """
        if self.model is None:
            print(" Error: Model not trained. Please run train_model() first.")
            return None

        # Convert to DataFrame if dict
        if isinstance(transaction_data, dict):
            transaction_data = pd.DataFrame([transaction_data])

        # Feature engineering
        df_pred = transaction_data.copy()
        df_pred['Is_Cross_Border'] = (df_pred['Sender_bank_location'] != df_pred['Receiver_bank_location']).astype(int)
        df_pred['Is_High_Risk_Location'] = df_pred.apply(
            lambda x: int(x['Sender_bank_location'] in ['AE-DXB', 'HK-HKG', 'RU-MOS'] or
                         x['Receiver_bank_location'] in ['AE-DXB', 'HK-HKG', 'RU-MOS']), axis=1
        )
        df_pred['Is_Structuring'] = df_pred['Amount'].apply(lambda x: int(9000 <= x <= 10000))
        df_pred['Currency_Match'] = (df_pred['Payment_currency'] == df_pred['Received_currency']).astype(int)
        df_pred['Hour'] = df_pred['Time'].apply(lambda x: int(x.split(':')[0]))
        df_pred['Is_Weekend'] = pd.to_datetime(df_pred['Date']).dt.dayofweek.isin([5, 6]).astype(int)
        df_pred['Is_Night_Time'] = df_pred['Hour'].apply(lambda x: int(x < 6 or x > 22))
        df_pred['Amount_Log'] = np.log1p(df_pred['Amount'])

        # Encode categorical variables
        for col in ['Payment_type', 'Payment_currency', 'Sender_bank_location', 'Receiver_bank_location']:
            if col in self.label_encoders:
                df_pred[f'{col}_encoded'] = self.label_encoders[col].transform(df_pred[col].astype(str))

        # Select features
        X_pred = df_pred[self.feature_names]
        X_pred_scaled = self.scaler.transform(X_pred)

        # Predict
        predictions = self.model.predict(X_pred_scaled)
        probabilities = self.model.predict_proba(X_pred_scaled)

        results = []
        for i, (pred, prob) in enumerate(zip(predictions, probabilities)):
            risk_level = 'HIGH RISK' if pred == 1 else 'LOW RISK'
            confidence = prob[pred] * 100

            results.append({
                'Transaction_Index': i,
                'Prediction': risk_level,
                'Confidence': confidence,
                'Laundering_Probability': prob[1] * 100,
                'Normal_Probability': prob[0] * 100,
                'Requires_Review': pred == 1
            })

        return pd.DataFrame(results)

    # ==================== COMPREHENSIVE REPORT ====================

    def generate_comprehensive_report(self, output_file='AML_Compliance_Report.txt'):
        """Generate a comprehensive AML compliance report"""
        print("\n" + "="*80)
        print("GENERATING COMPREHENSIVE AML COMPLIANCE REPORT")
        print("="*80)

        report = []
        report.append("="*80)
        report.append("Fraud Management AI SYSTEM - COMPREHENSIVE ANALYSIS REPORT")
        report.append("="*80)
        report.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        report.append("="*80)

        # Dataset Overview
        report.append("\n" + "="*80)
        report.append("1. DATASET OVERVIEW")
        report.append("="*80)
        report.append(f"Total Transactions: {len(self.df):,}")
        report.append(f"Date Range: {self.df['Date'].min()} to {self.df['Date'].max()}")
        report.append(f"Suspicious Transactions: {self.df['Is_laundering'].sum():,} ({self.df['Is_laundering'].sum()/len(self.df)*100:.2f}%)")
        report.append(f"Total Transaction Volume: ${self.df['Amount'].sum():,.2f}")
        report.append(f"Average Transaction Amount: ${self.df['Amount'].mean():,.2f}")

        # Customer Profiling Summary
        if self.customer_profiles is not None:
            report.append("\n" + "="*80)
            report.append("2. CUSTOMER PROFILING SUMMARY")
            report.append("="*80)
            report.append(f"Total Customers Profiled: {len(self.customer_profiles):,}")

            risk_dist = self.customer_profiles['Risk_Category'].value_counts()
            report.append("\nRisk Distribution:")
            for category, count in risk_dist.items():
                percentage = (count / len(self.customer_profiles)) * 100
                report.append(f"  {category:8s}: {count:4d} ({percentage:5.2f}%)")

            high_risk = self.customer_profiles[self.customer_profiles['Risk_Category'] == 'High']
            report.append(f"\nTop 5 High-Risk Customers:")
            for _, customer in high_risk.head(5).iterrows():
                report.append(f"  {customer['Account']}: Risk Score {customer['Risk_Score']}, "
                            f"{customer['Suspicious_Count']} suspicious transactions")

        # Anomaly Detection Summary
        if self.anomalies is not None:
            report.append("\n" + "="*80)
            report.append("3. ANOMALY DETECTION SUMMARY")
            report.append("="*80)
            report.append(f"Total Anomalies Detected: {len(self.anomalies):,}")
            report.append(f"Critical Priority: {(self.anomalies['Priority'] == 'CRITICAL').sum():,}")
            report.append(f"High Priority: {(self.anomalies['Priority'] == 'HIGH').sum():,}")
            report.append(f"Requires Immediate Review: {self.anomalies['Requires_Review'].sum():,}")

            report.append(f"\nTop Anomaly Flags:")
            all_flags = []
            for flags_str in self.anomalies['Flags'].dropna():
                all_flags.extend(flags_str.split(','))
            flag_counts = Counter(all_flags).most_common(5)
            for flag, count in flag_counts:
                report.append(f"  {flag}: {count:,}")

        # Policy Compliance Summary
        if self.policy_violations is not None and len(self.policy_violations) > 0:
            report.append("\n" + "="*80)
            report.append("4. POLICY COMPLIANCE SUMMARY")
            report.append("="*80)
            report.append(f"Total Policy Violations: {len(self.policy_violations):,}")

            severity_dist = self.policy_violations['Severity'].value_counts()
            report.append("\nViolation Severity:")
            for severity, count in severity_dist.items():
                report.append(f"  {severity}: {count:,}")

            report.append("\nTop Violated Regulations:")
            reg_dist = self.policy_violations['Regulation'].value_counts()
            for reg, count in reg_dist.head(5).items():
                report.append(f"  {reg[:60]}: {count:,}")

        # Model Performance Summary
        if self.model_metrics:
            report.append("\n" + "="*80)
            report.append("5. MODEL PERFORMANCE SUMMARY")
            report.append("="*80)
            report.append(f"Accuracy:  {self.model_metrics['accuracy']:.2f}%")
            report.append(f"Precision: {self.model_metrics['precision']:.2f}%")
            report.append(f"Recall:    {self.model_metrics['recall']:.2f}%")
            report.append(f"F1-Score:  {self.model_metrics['f1_score']:.2f}%")
            report.append(f"ROC-AUC:   {self.model_metrics['roc_auc']:.4f}")

            report.append("\nTop 5 Important Features:")
            feature_imp_sorted = sorted(
                self.model_metrics['feature_importance'].items(),
                key=lambda x: x[1],
                reverse=True
            )
            for feature, importance in feature_imp_sorted[:5]:
                report.append(f"  {feature}: {importance:.4f}")

        # Recommendations
        report.append("\n" + "="*80)
        report.append("6. RECOMMENDATIONS")
        report.append("="*80)
        if self.customer_profiles is not None:
            high_risk_count = len(self.customer_profiles[self.customer_profiles['Risk_Category'] == 'High'])
            if high_risk_count > 0:
                report.append(f"‚Ä¢ Immediate enhanced due diligence required for {high_risk_count} high-risk customers")

        if self.anomalies is not None:
            critical_count = (self.anomalies['Priority'] == 'CRITICAL').sum()
            if critical_count > 0:
                report.append(f"‚Ä¢ File Suspicious Activity Reports (SARs) for {critical_count} critical anomalies")

        if self.policy_violations is not None and len(self.policy_violations) > 0:
            critical_violations = len(self.policy_violations[self.policy_violations['Severity'] == 'Critical'])
            if critical_violations > 0:
                report.append(f"‚Ä¢ Address {critical_violations} critical policy violations immediately")

        report.append("‚Ä¢ Implement continuous monitoring for flagged accounts")
        report.append("‚Ä¢ Update risk models monthly with new transaction data")
        report.append("‚Ä¢ Conduct quarterly compliance audits")

        report.append("\n" + "="*80)
        report.append("END OF REPORT")
        report.append("="*80)

        # Save report
        report_text = '\n'.join(report)
        with open(output_file, 'w') as f:
            f.write(report_text)

        print(report_text)
        print(f"\n--> Report saved to '{output_file}'")

        return report_text

    # ==================== RUN ALL MODULES ====================

    def run_complete_analysis(self):
        """Run all modules in sequence"""
        print("\n" + " --- "*40)
        print("STARTING COMPLETE AML COMPLIANCE ANALYSIS")
        print("----- "*40)

        # Module 1: Customer Profiling
        self.customer_profiling()

        # Module 2: Anomaly Detection
        self.anomaly_detection()

        # Module 3: Policy Compliance
        self.policy_compliance_check()

        # Module 4: Model Training
        self.train_model()

        # Generate Report
        self.generate_comprehensive_report()

        print("\n" + "-->"*40)
        print("COMPLETE Fraud Detect ANALYSIS FINISHED")
        print("-->"*40)
        print("\nAll results saved. Check the generated files:")
        print("  ‚Ä¢ customer_risk_profiles.csv")
        print("  ‚Ä¢ detected_anomalies.csv")
        print("  ‚Ä¢ policy_violations.csv")
        print("  ‚Ä¢ aml_model_rf.pkl")
        print("  ‚Ä¢ AML_Compliance_Report.txt")
        print("  ‚Ä¢ *.png (visualizations)")


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================

if __name__ == "__main__":
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                                        ‚ïë
    ‚ïë            FRAUD DETECT AI SYSTEM - PYTHON IMPLEMENTATION            ‚ïë
    ‚ïë                                                                        ‚ïë
    ‚ïë  Complete Anti-Money Laundering Detection & Risk Assessment Platform  ‚ïë
    ‚ïë                                                                        ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)

    # Initialize system
    aml_system = AMLComplianceSystem()

    # Load data (use Google Drive link or local file)
    # For Google Drive: data_path = 'https://drive.google.com/file/d/1AxCyxmfbAMgPMhxQAnyWcieCaDMfhCMQ/view?usp=sharing'
    # For local file: data_path = 'transactions.csv'

    data_path = 'https://drive.google.com/file/d/1AxCyxmfbAMgPMhxQAnyWcieCaDMfhCMQ/view?usp=sharing'
    aml_system.load_data(data_path)

    # Run complete analysis
    aml_system.run_complete_analysis()

    # Example: Predict risk for a new transaction
    print("\n" + "="*80)
    print("EXAMPLE: PREDICTING RISK FOR NEW TRANSACTION")
    print("="*80)

    new_transaction = {
        'Time': '14:30:00',
        'Date': '2024-06-15',
        'Sender_account': 'ACC0001',
        'Receiver_account': 'ACC0002',
        'Amount': 9500,
        'Payment_currency': 'USD',
        'Received_currency': 'USD',
        'Sender_bank_location': 'US-NY',
        'Receiver_bank_location': 'AE-DXB',
        'Payment_type': 'Wire'
    }

    prediction = aml_system.predict_compliance_risk(new_transaction)
    print("\nPrediction Result:")
    print(prediction)

    print("\n *** Fraud Detect Analysis Complete!")
    print("=" * 80)
````
